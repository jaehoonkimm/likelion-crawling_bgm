# Function

```python
def get_items(keyword):
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
    params = {"items": "search_type%3Dall%26limit%3D40%26digrank%3D280%26reqtags%3Dremix%26lic%3Dall%26searchp%3D"+keyword+"%26dataview%3Dlinks_by%26cache%3D_593956693%26f%3Djsex", 
              "total": "f%3Dcount%26search_type%3Dall%26reqtags%3Dremix%26lic%3Dall%26searchp%3D"+keyword+"%26dataview%3Dlinks_by",
              "artists": "limit%3D40%26remixmin%3D1%26searchp%3D"+keyword+"%26dataview%3Duser_basic%26f%3Djsex"
             }
    r = requests.get(url, params=params, headers=headers)
    return json.loads(r.text)

def get_mp3_urls(body):
    
    temp = []
    for i in body[0]['items']:
        a = [j['download_url'] for j in i['files'] if '.mp3' in j['download_url']]
        temp.append(a[0])
        
    return temp
```

```python
body = get_items("song")
```
